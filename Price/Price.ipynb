```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# Configurando variavel para abrir navegador
web = webdriver.Chrome()
# Abrindo google
web.get("https://www.google.com/webhp?hl=pt-BR&sa=X&ved=0ahUKEwjlyubZl7r2AhVIGbkGHdIxC4YQPAgI")
# Clicando no campo pesquisar do google ( Para pegar xpath entrar no modo dev do navegador e copiar pela propria tag)
web.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação do dolar")
# Pesquisar cotação
web.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
# Pegando cotação e salvando na variavel ct_dolar
ct_dolar = web.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')


web.get("https://www.google.com/webhp?hl=pt-BR&sa=X&ved=0ahUKEwjlyubZl7r2AhVIGbkGHdIxC4YQPAgI")
# Clicando no campo pesquisar do google ( Para pegar xpath entrar no modo dev do navegador e copiar pela propria tag)
web.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação do euro")
# Pesquisar cotação
web.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
# Pegando cotação e salvando na variavel ct_euro
ct_euro = web.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

web.get("https://www.melhorcambio.com/ouro-hoje")

# Pegando cotação e salvando na variavel ct_ouro
ct_ouro = web.find_element(By.XPATH,'//*[@id="comercial"]').get_attribute('value')
ct_ouro = ct_ouro.replace(",",".")


print("O valor atual do dolar é:", ct_dolar)
print("O valor atual do euro é:", ct_euro)
print("O valor atual do ouro é: ", ct_ouro)
web.quit()
```

```python
import pandas as pd
tab = pd.read_excel("Produtos.xlsx")
display(tab)
```

```python
tab.loc[tab["Moeda"]=="Dólar","Cotação"] = float (ct_dolar)
display(tab)
```

```python
tab.loc[tab["Moeda"]=="Euro","Cotação"] = float (ct_euro)
display(tab)
```

```python
tab.loc[tab["Moeda"]=="Ouro","Cotação"] = float(ct_ouro)
display(tab)
```

```python
tab["Preço de Compra"] = tab["Preço Original"]*tab["Cotação"]
display(tab)
```

```python
tab["Preço de Venda"] = tab["Preço de Compra"] * tab["Margem"]
display(tab)
```

```python
tab.to_excel("ProdutosNovo.xlsx",index=False)
```
